```python
from collections import defaultdict
import math
```

```python
x = {'a': 0, 'b': 1, 'c': 2, 'd': 3, 'e': 4, 'f': 5,
     'g': 6, 'h': 7, 'i': 8, 'j': 9, 'k': 10, 'l': 11,
     'm': 12,'n': 13, 'o': 14, 'p': 15, 'q': 16, 'r': 17,
     's': 18, 't': 19, 'u': 20, 'v': 21, 'w': 22, 'x': 23,
     'y': 24, 'z': 25}
mod = 26
text = 'jgtnwtbwdvffvayzmzbvg'
```

```python
def encry(x,a,b,mod,text):
  l = []
  for i in text:
    find_num = x[i]
    l.append(((a*find_num)+b) % mod)


  encry_text = ''
  for i in l:
    for j in x:
      if x[j] == i:
        encry_text += j
  return encry_text
```

```python
encry_text = encry(x,a,b,mod,text)
print(encry_text)
    
```

```python
def decry(x,a,b,mod,encry_text):

  a_inverse = 0
  for i in range(1,mod):
    if ((a*i) % mod) == 1:
      a_inverse = i

  decry_num = []
  for i in encry_text:
    decry_num.append((a_inverse*(x[i]-b)) % mod)

  decry_text = ''
  for i in decry_num:
    for j in x:
      if x[j] == i:
        decry_text += j
  return decry_text
```

```python
a_value = []
for i in range(mod):
  if math.gcd(i,mod) == 1:
    a_value.append(i)
```

```python
for a in a_value:
  for b in range(mod):
    decry_text = decry(x,a,b,mod,'jgtnwtbwdvffvayzmzbvg')
    print(a,b,decry_text)
```
